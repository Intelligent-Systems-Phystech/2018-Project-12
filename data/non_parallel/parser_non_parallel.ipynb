```python
import pandas as pd
import re
```

```python
data = pd.read_csv('uk.txt', sep = ' ', names=['words', 'freq'])
data = data.sort_values(by='freq');
freq_dict = list(data['words'][-5000:])
```

```python
data.head()
```

```python
res = []
res_copy = []
words = []
counter = 0
gen_num = 0
ru_range = 300000
ru_result = []
counter = 0
for z in range(ru_range):
    try:
        with open('ukr/'+str(z), 'r') as f:
            res_copy = f.read().splitlines()
        res_copy = [x for x in res_copy if x]
        for number, sentence in enumerate(res_copy):
            res.append(re.split('[^а-яА-ЯїіҐґЄєЇїІі]+', sentence))
        for num1, sen in enumerate(res):
            for num2, senn in enumerate(sen):
                sen[num2] = senn.lower()
        for number, x  in enumerate(res):
            for y in x:
                if y in freq_dict:
                    counter += 1
            if counter/len(x)>0.75:
                ru_result.append(res_copy[number])
            counter = 0
        res = []
        print(z, len(ru_result))
    except FileNotFoundError:
        continue
```
